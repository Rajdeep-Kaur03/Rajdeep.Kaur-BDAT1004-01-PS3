```python
#importing necessary libraries
import pandas as pd
import numpy as np
```

```python
#Import the dataset
wind=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data', delim_whitespace=True)
wind
```

```python
#created a new col name datetime
wind["datetime"] = pd.to_datetime(wind[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1))
#replacing the first 3 columns
wind.drop(columns=['Yr','Mo','Dy'], inplace= True)
wind
```

```python
wind['datetime'] =wind['datetime'].apply(lambda x:x.replace(x.year - 100) if x.year>=2000 else x)
wind
```

```python
#Set the right dates as the index
wind["datetime"] = pd.to_datetime(wind["datetime"])
wind=wind.set_index("datetime")
wind
```

```python
#computing the missing values
print("Missing values")
wind.isnull().sum()
```

```python
#dropping the empty values and computing the non missing values 
wind = wind.dropna()
sum(wind.count())
```

```python
#calculate the mean windspeed
mean_windspeed=wind.mean(axis=0).mean()
print("The mean windspeeds of the windspeeds over all the locations and all the times:",mean_windspeed)
```

```python
#Creating a DataFrame and calculating the attributes usind describe function
loc_stats= wind.describe()
loc_stats
```

```python
#dropping count, 25%, 50% and 75%
loc_stats.drop(['count','25%','50%','75%'], inplace= True)
loc_stats
```

```python
#minimum windspeed across all the locations at each day
print("Minimum windspeed across all the locations at each day")
minimumvalue=wind[['RPT', 'VAL','ROS','KIL','SHA','BIR','DUB','CLA','MUL','CLO','BEL','MAL']].min(axis=1)
minimumvalue
```

```python
#maximum windspeed across all the locations at each day
print("Maximun windspeed across all the locations at each day")
maximumvalue=wind[['RPT', 'VAL','ROS','KIL','SHA','BIR','DUB','CLA','MUL','CLO','BEL','MAL']].max(axis=1)
maximumvalue
```

```python
#mean windspeed across all the locations at each day
print("Mean windspeed across all the locations at each day")
meanvalue=wind[['RPT', 'VAL','ROS','KIL','SHA','BIR','DUB','CLA','MUL','CLO','BEL','MAL']].mean(axis=1)
meanvalue
```

```python
#standard deviations of the windspeed across all the locations at each day
print("standard deviations of the windspeed across all the locations at each day")
standard_deviationValue=wind[['RPT', 'VAL','ROS','KIL','SHA','BIR','DUB','CLA','MUL','CLO','BEL','MAL']].std(axis=1)
standard_deviationValue
```

```python
#average windspeed in January for each location
avg_Jan = wind.groupby(wind.index.month).mean()
January= avg_Jan.head(1)
January
```

```python
#yearly frequency for each location
wind.groupby(wind.index.year).mean()
```

```python
#monthly frequency for each location
wind.groupby(wind.index.to_period('m')).mean()
```

```python
#weekly frequency for each location
wind.groupby(wind.index.to_period('w')).mean()
```

```python
#Calculate the minimum windspeeds across all locations for each week for the first 52 weeks.
minimumvalue=wind.groupby(wind.index.to_period('w')).min()
minimumvalue=minimumvalue.iloc[1:53]
minimumvalue
```

```python
#Calculate the maximum windspeeds across all locations for each week for the first 52 weeks.
maximumvalue=wind.groupby(wind.index.to_period('w')).max()
maximumvalue=maximumvalue.iloc[1:53]
maximumvalue
```

```python
#Calculate the mean windspeeds across all locations for each week for the first 52 weeks
meanvalue=wind.groupby(wind.index.to_period('w')).mean()
meanvalue=meanvalue.iloc[1:53]
meanvalue
```

```python
#Calculate the standard deviation windspeeds across all locations for each week for the first 52 weeks
standard_deviationvalue=wind.groupby(wind.index.to_period('w')).std()
standard_deviationvalue=standard_deviationvalue.iloc[1:53]
standard_deviationvalue
```

```python
#import the necessary libraries
import pandas as pd
import numpy as np
```

```python
#importing the dataset and assigning the name chipo
chipo=pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

```python
#first 10 entries
chipo.head(10)
```

```python
# number of observations in the dataset
num=chipo.index
print("Number of observations:", len(num))
```

```python
#number of columns in the dataset
print("Number of columns:",len(chipo.columns))
```

```python
#name of all the columns
print("column name:")
for col in chipo.columns: 
    print(col)
```

```python
# dataset index
chipo.index
```

```python
#most-ordered item
mostord= chipo.groupby(['item_name'],as_index=False).count()
mostord= mostord.sort_values(by='quantity', ascending=False)
mostord= mostord.head(1)
print("Most ordered item:",mostord['item_name'].to_string(index=False))
```

```python
#For the most-ordered item, no. of items ordered
print("Most ordered item = ",mostord['item_name'].to_string(index=False))
print('Number of items ordered:',mostord['quantity'].to_string(index=False))
```

```python
#most ordered item in the choice_description column
choice= chipo.groupby(['choice_description'],as_index=False).count()
choice= choice.sort_values(by='quantity', ascending=False)
choice = choice.head(1)
print("Most ordered choice description item is:",choice['choice_description'].to_string(index=False))
```

```python
#find the number of total items order
tot_ord = chipo.groupby(['item_name'],as_index=False).sum()
print("Total items ordered:",len(tot_ord))
```

```python
#Turn the item price into a float
chipo['item_price']=chipo.item_price.str.replace('$','')
chipo['item_price']=chipo['item_price'].astype(float)
chipo.dtypes
```
